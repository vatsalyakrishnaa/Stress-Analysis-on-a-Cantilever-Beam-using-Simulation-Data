```python
import pandas as pd

#load the excel file
data=pd.read_excel("beam_stress_data.xlsx")

#display the first 5 rows
data.head()
```

```python
import matplotlib.pyplot as plt

#plot stress vs load
plt.figure(figsize=(8,5))
plt.plot(data['Load (N)'], data['Stress (Pa)'], marker='*', color='m')
plt.title('Stress vs Load on Beam')
plt.xlabel('Load (N)')
plt.ylabel('Stress (Pa)')
plt.grid(True)
plt.show()
```

```python
#plot stress vs distance
plt.figure(figsize=(8,5))
plt.plot(data['Distance from Fixed End (m)'], data['Stress (Pa)'], marker='x', color='c')
plt.title('Stress vs Distance from fixed end')
plt.xlabel('Distance (m)')
plt.ylabel('Stress (pa)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load your Excel file
data = pd.read_excel("beam_stress_data.xlsx")

# Step 2: Add Moment and Stress columns
data['Moment (N*m)'] = data['Load (N)'] * data['Distance from Fixed End (m)']
data['Stress (Pa)'] = data['Moment (N*m)'] * 5000  # Example constant

# Step 3: Save updated file
data.to_excel("beam_stress_data_updated.xlsx", index=False)

# Step 4: Display data
print(data.head())
```
